Firstly we will implement linear k-means clustering algorithm.

```python
import matplotlib.pyplot as plt
import numpy as np
import torch
import time
from tqdm.notebook import trange, tqdm
from sklearn.datasets import fetch_openml

X, y = fetch_openml("mnist_784",
                   return_X_y=True)
# Convert data to image pixel-by-pixel representation
X_images = X.to_numpy().reshape(X.shape[0], 28, 28)

# Flatten the data so that we can apply clustering
X = X_images.reshape(X.shape[0], -1)
y = y.astype(int).to_numpy()

# Number of datapoints
num_img = 70000
# Number of cluster centers
num_means = 10
# Number of algorithm iterations
iterations = 100
# Size of each image (28*28 pixels)
data_size = 28*28

# Randomly generate K indices for K datapoints from the dataset
means = X[np.random.randint(0, num_img, num_means)]

# K-Means algorithm
start_time = time.time()

# Initialize a vector to contain the cluster index for each datapoint
cluster_index = np.zeros(num_img)

for i in trange(iterations, leave=False):
    # Assign each datapoint to the closest cluster
    for j in range(num_img):
        dist_to_center = [np.mean((X[j, :] - X[k, :])**2) for k in range(num_means)]
        cluster_index[j] = np.argmin(dist_to_center)

    # Update cluster centers
    for o in trange(num_means, leave=False):
        count = 0
        cluster_sum = 0
        for p in range(num_img):
            if cluster_index[p] == o:
                count += 1
                cluster_sum += X[p, :]
        if count > 0:
            means[o, :] = cluster_sum / count

end_time = time.time()
process_time = end_time - start_time
print("%d iterations took %.2f seconds, %.2fs/iteration" % (iterations,process_time, process_time / iterations))
```

Now we will perform optimized version of k-means clustering.

Parallelization is performed by GPU acceleration with PyTorch to speed up the k-means clustering algorithm, instead of CPU. Also optimized version utilizes torch for tensor operations instead of numpy arrays, matrix operations and broadcasting to calculate distances efficiently instead of nested loops, matrix multiplication and division to update cluster centers efficiently intead of nested loops.

```python
# Check for GPU availability
gpu_indx = 0
device = torch.device(gpu_indx if torch.cuda.is_available() else "cpu")
print("cuda or cpu?:", device)


# Rescale pixel values to be between 0-1
X = torch.FloatTensor((X.astype(float) / 255)).to(device)

# K-Means Initialization
means = X[np.random.randint(0, num_img, num_means)]
eye_mat = torch.eye(num_means, device=device)

# No-op
means = torch.mm(eye_mat, means)

# Visualize initial cluster centers
plt.figure(1, figsize=(20, 10))
img = means.cpu().float().numpy().reshape(num_means, 28, 28).transpose((1, 0, 2)).reshape(28, num_means*28)
_ = plt.imshow(img)

# K-Means algorithm with GPU acceleration
start_time = time.time()
with torch.no_grad():
    for i in trange(iterations):
        # Add a dimension and use broadcasting to find differences
        diff_from_means = means.unsqueeze(0) - X.unsqueeze(1)
        # Calculate absolute sum of differences
        dist_to_means = diff_from_means.pow(2).mean(2)
        # Find the index of the closest mean for each datapoint
        indx_of_means = dist_to_means.argmin(1)
        # Create a one-hot coded vector per datapoint
        a = eye_mat[indx_of_means].t()
        # Update cluster centers using matrix multiplication and division
        means = torch.mm(a, X) / a.sum(1, keepdims=True)

end_time = time.time()
par_process_time = end_time - start_time
print("%d iterations took %.2f seconds, %.4fs/iteration" % (iterations, par_process_time, par_process_time/iterations))

```

```python
speedup = (process_time/iterations) / (par_process_time/iterations)
round(speedup, 3)
```

```python
methods = ['Linear version', 'Optimized version']
time = [process_time/iterations, par_process_time/iterations]

plt.bar(methods, time)
plt.title('Speedup')
plt.xlabel('Versions')
plt.ylabel('Time')
plt.show()
```
